# Data Cleaning using Pandas

 Removing the un necessary data from cars fields

```python
tech['cars']
```

```python
tech['cars']=tech['cars'].str.upper()
tech['cars']
```

```python
tech[['h','names','bname','c','d','e','f','g','h','i']] =tech['cars'].str.split(' ',expand=True)
tech
```

```python
tech.drop(['h','e','d','f','g','i'],axis=1,inplace=True)
tech.head()
```

```python
tech['carName']=tech['names']+tech['bname']+' '+tech['c']
tech
```

```python
tech.drop(['names','bname','c','cars'],axis=1,inplace=True)
```

```python
tech.head(600)
```

```python
title = list(tech.columns)
title
```

```python
title[0],title[4]=title[4],title[0]
title
```

```python
title[1],title[4]=title[4],title[1]
title
```

```python
tech=tech[title]
tech
```

```python
tech.info()
```

# Visualisation
 Cars count in Differnt Cities

```python
ax=tech['city'].value_counts().plot(kind='pie')
ax.set_title('Number of cars in each city',fontweight='bold',fontsize=20)
```

```python
plt.figure(figsize=(16, 6))
t=sns.countplot(x='city',data=tech,palette='viridis')
b=t.set_xticklabels(t.get_xticklabels(), rotation=40, ha="right")
b=plt.title("Cars for sale in cities",fontsize=20,fontweight="bold")
plt.tight_layout()
plt.show(b)
```
